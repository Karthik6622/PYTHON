```python
#import csv module
import csv
#giving the column value in list
field=['S.no','NAME','Branch','Year']
#creating the list for row values
rows=[[1,'Parthi','CSE',2020],[1,'Parthi','CSE',2020],[1,'Parthi','CSE',2020],[1,'Parthi','CSE',2020]]
with open("E:\PYTHON\FILE\kar1.csv",'w')as f:
    #create the csv writer
    writer=csv.writer(f)
    #writing the fields in to the csv
    writer.writerow(field)
    #writing the row values in to the csv file
    writer.writerows(rows)
```

```python
#writing the csv file using dictionary
import csv
f=['Name','Branch','ID']
d=[{'Branch':'CSE','Name':'kani','ID':1234},{'Branch':'ECE','Name':'karthi','ID':12345}]
with open('E:\PYTHON\FILE\kar2.csv','w')as file:
    writer=csv.DictWriter(file,f)
    writer.writeheader()
    writer.writerows(d)
```

```python
x=bool([])#this one convert list value into boolean value,if we give empty list,it will give result as False.if we give list with values,it will give true value
x1=bool([5,9])
print(x,x1)
print('-'*50)
y=all([])#it will check whether we given all condition is True or not. if all condition are true,it will give result as True.if one or all condition are fail,it will give result as false
y1=all([5,'kk'])
y2=all([10==10,5==6])
print(y,y1,y2)
print('-'*50)
z=any([])#it will check whether we given one condition is true or not,if one condition is true, it give result as True.if all conditions are fail,it will give result as false.
z1=any([3,4,5])
z2=any([3<9,89>0,0>0])
z3=any([3>9,89<0,0>9])
print(z,z1,z2,z3)
print('-'*50)
print(x==z and y)



```

```python
#import CSV MOdules and writes data into the file
import csv
field=['NAME','AGE','GENDER','CITY','Degree']
datas=[['Karthik',23,'Male','Tcr'],['Mani',23,'Male','Tcr'],['Raman',23,'Male','Tcr']]
with open('E:/PYTHON/FILE/CSV_FILE_HANDLING1.csv','w')as f:
    wr=csv.writer(f)
    wr.writerow(field)
    wr.writerows(datas)
    f.close()
```

```python
#write dictionary data into file using csv modules
import csv
fields=['NAME','Accountnumber','BIN','ADDRESS']
datas=[{'NAME':'KARTHIK','Accountnumber':909090,'BIN':9999,'ADDRESS':'Tiruchendur'},{'NAME':'MANI','Accountnumber':786590,'BIN':0000,'ADDRESS':'Tiruchendur'}]#hhat we have given in fields should be same as here
with open('E:/PYTHON/FILE/CSV_FILE_HANDLING2.csv','w')as k:
        wr=csv.DictWriter(k,fields)
        wr.writeheader()
        wr.writerows(datas)
     
```
