```python
'''
1) Hello World – Skip repeated element that should be 
show in one time and display the result
'''
s='Hello World'
li=list(s)
#print(li)
temp=[]
for i in li:
    
    if i not in temp:
        temp.append(i)
            
print(''.join(temp))            
            
```

```python
'''
2) Calendar - Previous 10 years calendar – 2023 – From 
current month to 10 years
'''
#find out sep-2023 to sep-2013
import calendar
import datetime
today=datetime.datetime.now()
current_month=today.month
current_year=today.year
for i in range(current_year,current_year-11,-1):
    if (i==current_year):
        for j in range(current_month,13):
            print(calendar.month(i,j))

    elif (i==current_year-10):
        for j in range(1,current_month+1):
            print(calendar.month(i,j))
    else:
        for j in range(1,13):
            print(calendar.month(i,j))
```

```python
#3) List – Maximum repeated element two numbers.
from collections import Counter
li=[45,6,33,0,56,9,45,45,89,0,56,9]
c=Counter(li)
most=c.most_common(2)
for value,count in most:
    print(f'Value:{value} count:{count}')
```

```python
'''
4) List – Ascending and descending order without using 
sort and reverse attribute
'''
li=[45,6,33,0,56,9]
length=len(li)
for i in range(length):
    for j in range(length):
        if (li[i]<li[j]):
            temp=li[i]
            li[i]=li[j]
            li[j]=temp
print("Accending order:",li)
li=[45,6,33,0,56,9]
length=len(li)
for k in range(length):
    for l in range(length):
        if (li[k]>li[l]):
            temp=li[k]
            li[k]=li[l]
            li[l]=temp
print("Decending order:",li)
```
