# Simple Linear Regression Using Student Data

# Importing Libraries

```python
import numpy as np
import pandas as pd
```

# Load The Dataset

```python
data=pd.read_csv('Students.csv')
data
```

# summarize the Dataset


```python
data.shape#30-rows,2-columns
```

```python
data.head()
```

# Segregateb Dataset into X(Input/Independent Variable) & Y(Output/Dependent Variable)

```python
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
X
```

```python
Y
```

# Spliting the Dataset into Train and Test

```python
#sklearn-scikitlearn
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=1)#it's like a seed once value assigend,it won't be change
```

```python
X_train
```

```python
X_test
```

# Training the data (Chooing the Model (Algorithm))

```python
from sklearn.linear_model import LinearRegression
stud=LinearRegression()
stud.fit(X_train,Y_train)
```

# Prediction For all Test Data

```python
Y_predict=stud.predict(X_test)
```

```python
Y_test
```

```python
Y_predict
```

# plotting

```python
import matplotlib.pyplot as plt
plt.scatter(X_test,Y_test)
plt.plot(X_test,Y_predict,'ro:')
plt.title('Hours of Learning VS Marks')
plt.xlabel('Hours of Learning')
plt.ylabel('Marks')
plt.show()
```
