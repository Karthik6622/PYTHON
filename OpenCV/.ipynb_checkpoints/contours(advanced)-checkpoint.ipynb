# Contours in opencv

Contours in OpenCV are a fundamental concept used in computer vision and image processing to identify and extract the boundaries of objects within an image. They are especially useful for tasks like object detection, shape analysis

```python
import cv2#importing opencv
```

```python
#fetching image from system
img=cv2.imread("groupofpeople.png")
#changing the image to grayscale becuse if image in grayscale ,easy to find out the contours or boundriedsof the objects
Gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#displaying both images color and grayscale
cv2.imshow("Normal Image",img)
cv2.imshow("Grayscale Image",Gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#now we are are going threshold the image object or cut the object in image
_,thresh=cv2.threshold(Gray_image,220,255,0)#here if pixel is lessthan 210 unchanged image will be displayed but if it is morethan 210,0 pixel(white) will be displayed
cv2.imshow("thresh Image",thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#create contours from thresh image
contour,hiracy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)#cv2.RETR_TREE: This mode retrieves all of the contours and reconstructs a full hierarchy of nested contours. It creates a tree-like structure that represents the parent-child relationships between contours. In this hierarchy:
print(contour)#displying contours x and y coortinates
'''
contours: This is a list that will store the contours found in the binary image. After running the cv2.findContours function, contours will contain all the contours detected in the image.

hierarchy: This is another list that stores the hierarchy information of the contours. The hierarchy describes the relationship between different contours when they are nested or connected in some way. 
'''
'''
cv2.RETR_EXTERNAL: This is a retrieval mode argument. It specifies the retrieval mode of the contours. In this case, cv2.RETR_EXTERNAL retrieves only the external contours, which are the contours of the top-level objects in the image. Any contours that are enclosed within other contours will be ignored.

cv2.CHAIN_APPROX_SIMPLE: This is a contour approximation method. It specifies how the contours are represented. cv2.CHAIN_APPROX_SIMPLE compresses horizontal, vertical, and diagonal segments and leaves only their end points. This reduces the number of points in the contour representation, making it more memory-efficient. If you need all the points of the contour, you can use cv2.CHAIN_APPROX_NONE.
'''
```

```python
#draw the contours on the original image
conn=cv2.drawContours(img,contour,-1,(0,255,0),3)
#img=original image
#-1=-1 means all contours will applay
#(0,255,0)-color
#3=size of line
cv2.imshow("contourImage",conn)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
ball_image=cv2.imread("color_circle.jfif")
plt.imshow(ball_image)#displaing image as a rgb format 
```

```python
#converting image as a bgr format
imgbgr=cv2.cvtColor(ball_image,cv2.COLOR_RGB2BGR)
plt.imshow(imgbgr)
```

```python
#changing image to contour image
img_grayscale=cv2.cvtColor(imgbgr,cv2.COLOR_BGR2GRAY)
_,thresh=cv2.threshold(img_grayscale,200,255,0)
contour,hiracy=cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
#cv2.RETR_LIST-it's give contours coortinates in list format
draw_contour=cv2.drawContours(imgbgr,contour,-1,(0,255,0),2)
plt.subplot(3,1,1)
plt.title("Gray scale")
plt.xticks([])
plt.yticks([])
plt.imshow(img_grayscale)
plt.subplot(3,1,2)
plt.title("Thresh image")
plt.xticks([])
plt.yticks([])
plt.imshow(thresh)
plt.subplot(3,1,3)
plt.title("Contour image")
plt.xticks([])
plt.yticks([])
plt.imshow(draw_contour)
```

```python
#creating simple motion sensor
#capturing video from our pc
ca=cv2.VideoCapture("mixkit-conceptual-video-of-businessman-walking-on-the-road-33034-medium.mp4")
#creating two frames
check,frame1=ca.read()
check,frame2=ca.read()
while (ca.isOpened()):
    #finding different between frame1 and frame2
    different=cv2.absdiff(frame1,frame2)
    #change diifernt to gray scale image
    gray=cv2.cvtColor(different,cv2.COLOR_BGR2GRAY)
    #blurring the image using Gaussian blur
    blur=cv2.GaussianBlur(gray,(5,5),0)
    #using threshold to seperate the image
    _,thresh=cv2.threshold(blur,50,255,cv2.THRESH_BINARY)
    #for fill the threshold image we using dilate function
    dilated=cv2.dilate(thresh,None,iterations=5)#it's fill darkness as a iteration value
    #creating the contours for thresh image
    contours,_=cv2.findContours(dilated,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
    #draw the contours
    #cont=cv2.drawContours(frame1,contours,-1,(0,255,0),3)
    #we going to draw the rectangle for contours
    for contour in contours:
        (x,y,w,h)=cv2.boundingRect(contour)#we changing x and y coordinates to rectangle (x,y,width,height)
        if cv2.contourArea(contour)<2000:#here we getting contour area size if area size is lessthan 2000,contour rectangle won't display
            continue
        cv2.rectangle(frame1,(x,y),(x+w,y+h),(0,255,0),5)#draw the rectangle 
        cv2.putText(frame1,"Movement",(150,20),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)#put the text on image )if the happening in the video,it will display the "movement" message
    cv2.imshow("video",frame1)
    if cv2.waitKey(50)==ord('q'):
        break
    frame1=frame2
    check,frame2=ca.read()

cv2.destroyAllWindows()
ca.release()
```

```python
#how to detect simple geometrical shapes using opencv
#read the image from pc
img=cv2.imread("shapes.jfif")
#changing image to grayscale
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#seperating thresh from the gray image
_,thresh=cv2.threshold(gray_img,70,255,0)
#getting contour from the thresh
contours,hieracy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
for contour in contours:
    #finding curve in the contour
    approx=cv2.approxPolyDP(contour,0.01*cv2.arcLength(contour,True),True)#true-True, indicates that the input contour is a closed contour, 
    #is using the cv2.approxPolyDP() function to approximate a contour (contour) by reducing the number of vertices based on a specified tolerance level.
    #cv2.arcLength(), in OpenCV is used to calculate the perimeter or the length of a closed contour or a curve. It's particularly useful in various computer vision and image processing task
    '''
    0.01 * cv2.arcLength(contour, True) is calculating 1% of the arc length of the contour. This value is often used as the tolerance or epsilon parameter in functions like cv2.approxPolyDP() to control the level of simplification when approximating a contour with a polygon.
    '''
    cv2.drawContours(img,[approx],-1,(0,255,0),3)
    #print(approx)
    #ravel()-It is used to flatten multi-dimensional arrays into a one-dimensional array. 
    x=approx.ravel()[0]
    y=approx.ravel()[1]-15
    #displaying the shapes names using thier curve or arclength
    if len(approx)==3:
        cv2.putText(img,"TRIANGLE",(x,y),cv2.FONT_HERSHEY_COMPLEX,0.4,(255,255,255),1)
    elif len(approx)==4:
        #here both rectangle and squre has same 4 arc lenth so how to differencies both of that
        x1,y1,w,h=cv2.boundingRect(approx)
        aspectratio=float(w)/h
        if aspectratio >=0.95 and aspectratio<=1.05:
             cv2.putText(img,"SQURE",(x,y),cv2.FONT_HERSHEY_COMPLEX,0.4,(255,255,255),1)
        else:
            cv2.putText(img,"RECTANGLE",(x,y),cv2.FONT_HERSHEY_COMPLEX,0.4,(255,255,255),1)
    elif len(approx)==10:
        cv2.putText(img,"STAR",(x,y),cv2.FONT_HERSHEY_COMPLEX,0.4,(255,255,255),1)
    elif len(approx)==7:
        cv2.putText(img,"HEPTAGON",(x,y),cv2.FONT_HERSHEY_COMPLEX,0.4,(255,255,255),1)
    elif len(approx)==6:
        cv2.putText(img,"HEXAGON",(x,y),cv2.FONT_HERSHEY_COMPLEX,0.4,(255,255,255),1)
    else:
        cv2.putText(img,"CIRCLE",(x,y),cv2.FONT_HERSHEY_COMPLEX,0.4,(255,255,255),1)
cv2.imshow("Normal Image",img)
cv2.waitKey()
cv2.destroyAllWindows()
```
