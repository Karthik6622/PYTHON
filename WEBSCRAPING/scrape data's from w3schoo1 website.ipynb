```python
#importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
##finding the Available courses in W3schools
#sending request to the website
page=requests.get("https://www.w3schools.com/")
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
#taking data's by "ga-nav" class
res=soup.find_all(class_="ga-nav")
#removing all tags and displaying only contents
s=str(res)
soup1=BeautifulSoup(s,'lxml').get_text()
print(soup1)
```

```python
#importing pandas to change scraped data into dataset
import pandas as pd
df1=pd.DataFrame([soup1])
df1
```

```python
#split the data's using split function
df2=df1[0].str.split(',',expand=True)
df2
```

```python
#deleting first two columns
df2.drop(columns=[0,1],inplace=True)
```

```python
df2
```

```python
#renaming the columns
import numpy as np
arr=np.arange(1,37)
dic=dict(zip(df2.columns,arr))
print(dic)
df2.rename(dic,axis=1,inplace=True)
```

```python
df2
```

```python
#rotate the table
df3=df2.transpose()
df3
```

```python
#changing column name as 0 to "Available course"
df3.rename(columns={0:'Availble course'},inplace=True)
df3
```

```python
##scraping all html topics
from urllib.request import urlopen
```

```python
#send the request to below mentioned url and getting response from the url
url="https://www.w3schools.com/html/default.asp"
html=urlopen(url)
html
```

```python
#display all elemnts in above mentioned url webpage
soup2=BeautifulSoup(html,'lxml')
print(soup2.prettify())
```

```python
#scrap the data by id
ress=soup2.find_all(id='leftmenuinnerinner')
#cleaning the data as removing the tags,changing to list format
import re
listt=[]
for i in ress:
    s1=str(i)
    com=re.compile('<.*?>')
    clean=re.sub(com,'',s1)
    listt.append(clean)
print(listt)
print(type(listt))
print('*'*50)
for j in listt:
    print(j)
```

```python
#changing website data to dataset
df5=pd.DataFrame(listt)
df5
```

```python
#spliting the data's 
df5=df5[0].str.split('\n',expand=True)
df5
```

```python
#transpose the dataframe
df5=df5.transpose()
df5
```

```python
#remove index postitions 0,1,128,129
df5.drop(index=[0,1,128,129],inplace=True)
df5
```

```python
##scraping all css topics
#send the request to below mentioned url and getting response from the url
url="https://www.w3schools.com/css/default.asp"
html=urlopen(url)
html
```

```python
#display all elemnts in above mentioned url webpage
soup2=BeautifulSoup(html,'lxml')
print(soup2.prettify())
```

```python
#changing website data to dataset
df6=pd.DataFrame(listt1)
df6
```

```python
#spliting the data's 
df6=df6[0].str.split('\n',expand=True)
df6
```

```python
#transpose the dataframe
df6=df6.transpose()
df6
```

```python
#delete index positions 0,1,207,208
df6.drop(index=[0,1,207,208],inplace=True)
df6
```

```python
df6.rename(columns={0:'CSS ALL TOPICS'},inplace=True)
df6
```

```python
#concating df5(htmlalltopicdataframe) and df6(cssalltopicdataframe) as a column
df7=pd.concat([df5,df6],axis=1)
df7
```
