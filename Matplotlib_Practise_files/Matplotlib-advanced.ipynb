# Matplotlib-advanced

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df=pd.read_excel('Covid cases in India.xlsx')
df
```

```python
style.use('classic')#using style.use() function we can set various style of our plot
plt.plot(df['Name of State / UT'])
```

```python
style.use('ggplot')
plt.plot(df['Name of State / UT'])
```

```python
#different types of style.use
'''
'default': This is the default Matplotlib style, which is simple and provides basic plot elements.

'classic': A style reminiscent of the classic Matplotlib 1.4 and earlier appearance.

'seaborn': A style similar to the Seaborn library's default style.

'ggplot': As mentioned earlier, this style mimics the aesthetics of the ggplot2 library in R.

'fivethirtyeight': Emulates the style of plots seen on the FiveThirtyEight website.

'bmh': Inspired by the Bayesian Methods for Hackers style.

'grayscale': A black-and-white style for simple and clean plots.

'tableau-colorblind10': A style that's designed to be friendly to colorblind individuals, using distinct colors for lines and markers.

'dark_background': A style with a dark background, suitable for presentations or when you want a high contrast appearance.
'''
```

# Data Visulization

```python
#usually we use plt.bar(),plt.pie(),plt.scatter()... to draw a plotchart or barchart..
#but here we use kind to draw we want plot
df2.plot(kind='bar',x='Name of State / UT',y='Cured')
```

```python
#kind='scatter'
df2.plot(kind='scatter',x='Name of State / UT',y='Cured')
```

```python
#normal plot
df2.plot(x='Name of State / UT',y='Cured')
```

```python
#kind=pie
df2.plot(kind='pie',y='Cured',labels=df2['Name of State / UT'])
plt.legend(df2['Name of State / UT'],loc=(1,0))
```

```python
#instead of x,y we can directly give the column names
plt.bar(df2['Name of State / UT'],df2['Death'])
```
