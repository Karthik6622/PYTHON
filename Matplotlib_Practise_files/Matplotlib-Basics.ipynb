# Matplotlib-Basics

```python
#how to check matplot version
import matplotlib
print(matplotlib.__version__)
```

```python
#plot-it is uesd to draw line graph
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,6])
y=np.array([0,200])
plt.plot(x,y)#It will generate the line plot
#it automatically generate line as we given x and y value
```

```python
#plot line  color is Defaultly Blue,if we want change color as we want,use below code
print(plt.plot(x,y,'y'))#r-red,k-black,etc...
x1=np.array([0,10])
y1=np.array([0,100])
plt.plot(x1,y1,color='orange')
```

```python
#we can give hexcode to change the color
x=np.array([0,6])
y=np.array([0,200])
plt.plot(x,y,color='#0f0f0f')#black
```

```python
#without line
plt.plot(x,y,'*')
```

```python
#linestyle-ls=we can change line style using ls
plt.plot(x,y,ls='dotted')
'''
    Solid Line:
        '-' or 'solid': Creates a solid line (default linestyle).

    Dashed Line:
        '--' or 'dashed': Creates a dashed line.

    Dotted Line:
        ':' or 'dotted': Creates a dotted line.

    Dash-Dot Line:
        '-.' or 'dashdot': Creates a dash-dot line.
'''
```

```python
x=np.array([80,85,90,100,96,105,110,78,115,125,120])
y=np.array([240,250,270,290,280,330,310,300,320,260,340])
plt.plot(x,y,color='r',marker='*',ms=10)#marker-it is x and y joining point ,ms(markersize)-size of the marker
```

```python
'''
these much shapes we can give as a marker value:
'.': Point marker
',': Pixel marker
'o': Circle marker
'v': Downward-pointing triangle marker
'^': Upward-pointing triangle marker
'<': Left-pointing triangle marker
'>': Right-pointing triangle marker
'1': Tri-down marker
'2': Tri-up marker
'3': Tri-left marker
'4': Tri-right marker
's': Square marker
'p': Pentagon marker
'*': Star marker
'h': Hexagon1 marker
'H': Hexagon2 marker
'|': Vertical line marker
'_': Horizontal line marker
'''
x=np.array([80,85,90,100,96,105,110,78,115,125,120])
y=np.array([240,250,270,290,280,330,310,300,320,260,340])
plt.plot(x,y,color='gray',lw=10,marker='p',ms=15,mfc='pink')#lw(line width)-width of the Line,mfc(marker face color)-using this we can give marker color
```

```python
x=np.array([3,8,1,10])
plt.plot(x,'violet',lw=5,marker='o',ms=10,mfc='y')
plt.xlabel("Life")#it's name of the Y
plt.ylabel("Expense")#it's name of the X 
plt.title("Karthik Graph")#it's a heading of graph
```

```python
#changing lablels and title colors,size,font
x=np.array([3,8,1,10])
plt.plot(x,'violet',lw=5,marker='o',ms=10,mfc='y')
plt.xlabel("Life",{'fontfamily':'Arial','color':'blue','size':20})
plt.ylabel("Expense",{'fontfamily':'Times New Roman','color':'green','size':20})
plt.title("Karthik Graph",{'fontfamily':'verdana','color':'brown','size':25})
```

```python
#how to increse plot graph Size
x=np.array([3,8,1,10])
plt.figure(figsize=(10,5))#1stnumber=horizontalsize,2ndnumber=vertical size of the fgraph
plt.plot(x,'violet',lw=5,marker='o',ms=10,mfc='y')
plt.xlabel("Life",{'fontfamily':'Arial','color':'blue','size':20})
plt.ylabel("Expense",{'fontfamily':'Times New Roman','color':'green','size':20})
plt.title("Karthik Graph",{'fontfamily':'verdana','color':'brown','size':25})
```

```python
#create wave form using matplotlib
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
print(x)
print(y)
plt.plot(x,y,'y',lw=6,marker="^",ms=10,mfc='orange')
plt.xlabel('Avg values')
plt.ylabel('Sin values')
plt.title("Wave Form")
```

```python
#create wave form using matplotlib
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
print(x)
print(y)
plt.plot(x,y,'y',lw=6,marker="^",ms=10,mfc='orange')
plt.xlabel('Avg values')
plt.ylabel('Sin values')
plt.title("Wave Form")
```

```python
#barchart-it is used to draw graph like a bar shape
x=np.array([1,2,3,4])
y=np.array([20,40,60,80])
plt.bar(x,y)
plt.xlabel('Avg value')
plt.ylabel('life value')
plt.title('Span value')
```

```python
#how to give different colors of each bar
x=np.array([1,2,3,4])
y=np.array([20,40,60,80])
colors=['red','blue','pink','orange']
plt.bar(x,y,color=colors)
plt.xlabel('Avg value')
plt.ylabel('life value')
plt.title('Span value')
```

```python
#vertical Bar chart
x=np.array([1,4])
y=np.array([15,60])
plt.bar(x,y,color='pink',width=1)#for bar we can give color and width also
plt.xlabel('Average Value')
plt.ylabel('Life Value')
plt.title('Span Value')
```

```python
#horizontal bar chart
x=np.array([1,20])
y=np.array([1,90])
print(plt.barh(x,y,color='g',height=0.8))#with out print we can give and with print also we can give.#barh-horizontal bar
plt.xlabel('Average Value')
plt.ylabel('Life Value')
plt.title('Span Value')
```

```python
#how to draw two graphs in one form
x=np.array([1,4,8,12])
y=np.array([10,20,100,40])
plt.bar(x,y,color='r',width=1)
plt.plot(x,y,color='Yellow',lw=10,marker='s',ms=10,mfc='k')#s=squre marker
```

```python
#piechart-it is used to draw pie chart from we given values
y=np.array([30,80,10,5])
plt.pie(y)
```

```python
#piechart-it is used to draw pie chart from we given values
fruits=["Apple","pomegrante",'Mango','Banana']
values=[20,50,40,100]
plt.pie(values,labels=fruits)#with label attribute we can give name each wedge
```

```python
#defaultly it will give sepeare color for each wedge,if we want our own color,that also we can do here
y=np.array([10,40,20,80])
fruits=["Apple","pomegrante",'Mango','Banana']
values=[20,50,40,100]
colors1=['orange','red','yellow','green']
plt.pie(values,labels=fruits,colors=colors1)
```

```python
#explode attribute-it will seperate it we given wedge
y=np.array([10,40,20,80])
fruits=["Apple","pomegrante",'Mango','Banana']
values=[20,50,40,100]
colors1=['orange','red','yellow','green']
exp=[0,0.2,0,0]#here i exploded Pommegrante
plt.pie(y,labels=fruits,colors=colors1,explode=exp)
```

```python
#shadow-it's will applay shadow effect to given wedge
#radius-it is use to increase the radius size of pie chart
y=np.array([10,40,20,80])
fruits=["Apple","pomegrante",'Mango','Banana']
values=[20,50,40,100]
colors1=['orange','red','yellow','green']
exp=[0,0.2,0,0]#here i exploded Pommegrante
plt.pie(y,labels=fruits,colors=colors1,explode=exp,radius=2,shadow=True)#radius won't work if we give frame
```

```python
#frame-it will draw line around the pie chart
y=np.array([10,40,20,80])
fruits=["Apple","pomegrante",'Mango','Banana']
values=[20,50,40,100]
colors1=['orange','red','yellow','green']
exp=[0,0.2,0,0]#here i exploded Pommegrante
plt.pie(y,labels=fruits,colors=colors1,explode=exp,frame=True,shadow=True)
```

```python
#legend=it will give explanation each wedges
y=np.array([10,40,20,80])
fruits=["Apple","pomegrante",'Mango','Banana']
values=[20,50,40,100]
colors1=['orange','red','yellow','green']
exp=[0,0.2,0,0]#here i exploded Pommegrante
plt.pie(y,labels=fruits,colors=colors1,shadow=True)
plt.legend(title="FRUIT NAMES",loc=(1,0))#title=tiltle of legend,loc(loction)-where we need legend box
'''
'best': Automatically choose the best location for the legend to avoid overlapping with plot elements.
'upper right': Upper right corner of the plot.
'upper left': Upper left corner of the plot.
'lower right': Lower right corner of the plot.
'lower left': Lower left corner of the plot.
'right': Right side of the plot.
'center left': Left side of the plot, centered vertically.
'center right': Right side of the plot, centered vertically.
'lower center': Bottom center of the plot.
'upper center': Top center of the plot.
'center': Center of the plot.
Additionally, you can use numerical codes (0 to 10) to specify the position of the legend. 
The code 0 corresponds to 'best', 1 to 'upper right', 2 to 'upper left', and so on.
'''
```
