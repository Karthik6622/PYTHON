# APPlYING MULTI LINEAR ALGORITHM FOR IOB2022 STATEMENT

```python
import numpy as mp
import pandas as pd
```

```python
#read the dataset using pandas
dd=pd.read_excel("2022_IOB_STATEMENT.xls")
dd
```

```python
dd['Date (Value Date)']=dd['Date (Value Date)'].apply(lambda x:x.split('(')[0])
```

```python
dd
```

```python
dd['Date (Value Date)']=dd['Date (Value Date)'].apply(lambda x:x.split('-')[1:3])
```

```python
dd
```

```python
dd['Date (Value Date)']=dd['Date (Value Date)'].apply(lambda x:','.join(x))
```

```python
dd.info()
```

```python
dd['Debit(Rs)'].astype(float)
dd['Credit(Rs)'].astype(float)
```

```python
dd.groupby(['Date (Value Date)'])[['Debit(Rs)','Credit(Rs)']].sum()
```
