# image gradient in opencv

An image gradient is a directional change in the intensity or color in an image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Laplacian method-it's one type of method to change image gradient
#read the color image and convert into grayscale
image=cv2.imread("ronaldo.jfif",cv2.IMREAD_GRAYSCALE)
#finding image gradient using laplacian method
lep=cv2.Laplacian(image,cv2.CV_64F)
'''
cv2.Laplacian calculates the Laplacian gradient of the input image. The second argument (cv2.CV_64F) specifies the data type of the output image, which is a 64-bit floating-point number. Using this data type allows for negative values in the Laplacian result.
'''
#changing to absolute and int datatype
lep=np.uint8(np.absolute(lep))
#unit8-unsigned 8-bit integer(it's like a fixed integer we can't give morethan 255)
imgs=[image,lep]
titles=['Original image','Gradient image']
for i in range(len(imgs)):
    plt.subplot(1,2,i+1)
    plt.imshow(imgs[i])
    plt.title(titles[i])
plt.show()
```

```python
#sobelx-we can calculate the Sobel gradient in the X (horizontal) direction of an image using the Sobel function.
img=cv2.imread("image.jpg")
#finding image gradient using sobelx method
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0)#1-x,0-y here we doing horizontal directional gradient so we making x=1 and y=0
#changing to integer
sobelx=np.uint8(np.absolute(sobelx))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(sobelx)
plt.title("sobelx gradient image")
```

```python
#sobely-To calculate the Sobel gradient in the Y (vertical) direction of an image using OpenCV, you can use the Sobel function just like you would for the Sobel X gradient
img=cv2.imread("image.jpg")
#finding image gradient using sobelx method
sobely=cv2.Sobel(img,cv2.CV_64F,0,1)#1-y,0-x here we doing vertical directional gradient so we making x=0 and y=1
#changing to integer
sobely=np.uint8(np.absolute(sobely))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(sobely)
plt.title("sobelx gradient image")
```

```python
#how to combine both horizontal(sobelx) and vertical(sobely) gradient images
combine=cv2.bitwise_or(sobelx,sobely)
#using bitwise_or we can combine both images
cv2.imshow("combined image",combine)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#edge detector
#using the canny function we can find out the image edges
img=cv2.imread("ronaldo.jfif")
#finding image edges using canny function
edge_detect=cv2.Canny(img,100,200)#100-threshold value1,200-threshold value2
'''
It requires two threshold values (threshold1 and threshold2). Pixels with gradient values above threshold2 are considered strong edges, while pixels between threshold1 and threshold2 are considered weak edges.
You can experiment with these thresholds to adjust the sensitivity of edge detection.
'''
#it give output as only pixel edges from 100 to 200 and 200 above 
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(edge_detect)
plt.title("edge detecter image")
```
