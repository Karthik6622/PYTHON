```python
#zip()-it used to combine more values to one value
new_column=['Fuel Consumption','Cylinders',
            'Disp','HP','Weight','Accer','Year',
            'Origin','Model']
dic=dict(zip(car.columns,new_column))
dic
```

```python
#rename-it is used to rename the only column(not row) as we need
#car.rename(new_column,axis=1,inplace=True)#error-we can't give list 
car.rename(dic,axis=1,inplace=True)
```

```python
#Finding the Average Value for the Column Name 'Fuel Consumption'
print(car['Fuel Consumption'].mean())
#Finding the first 5 row Average Value for the Column Name 'Fuel Consumption'
print(car['Fuel Consumption'].head().mean())
```

```python
#Sum of the Value for the Column Name ('Weight')
print(car['Weight'].sum())
#Sum of the last 5 Value for the Column Name ('Weight')
print(car['Weight'].tail().sum())
```

```python
#Find out the unique value in the 'Cylinders' column
car['Cylinders'].unique()
```

```python
#groupby-it is used to group the Dataset as we need
cyl=car.groupby('Cylinders')
```

```python
cyl.mean()
```

```python
#groupby-it is used to group the Dataset as we need
#example here for 'Cylinders' column we finding  sum of 'Fuel consumption' column values
cyl1=car.groupby(['Cylinders'])['Fuel Consumption'].sum()
cyl1
```

```python
cyl1=car.groupby(['Origin'])['Fuel Consumption','Disp','HP'].sum()
cyl1
```

```python
#info()-it is used to give information about we given variable
car.info()
```

```python
#describe()-it is used to describe the varible
car.describe()#it will give count ,mean,std,min,max values
```

```python
#agg()-Aggregation functions perform calculations on sets of values, such as calculating sums, averages, counts, maximum values, minimum values, etc.
print(cyl.mean())
print('*'*50)
print(cyl.sum())
print('*'*50)
print(cyl.max())
print('*'*50)
def average(data):
    return data/len(data)
print(cyl.agg(average))#by agg function we can pass our user-defined function
print('*'*50)
cyl.agg(['mean','max','min','sum'])
```

```python
cyl['Fuel Consumption'].agg(['max','mean','count'])#finding cyl variable 'fuel consumption' column max,mean,count
```

```python
car['Fuel Consumption'].agg(['max','mean','count'])#finding car variable 'fuel consumption' column max,mean,count
```

```python
#apply- it is used to applay user defied function or Built-in function to datasets
def divide_by_two(data):
    return data/2.0
car['Accer'].apply(divide_by_two)
```

```python
df=pd.read_csv('E:/Python_notes/data_car_cross.csv',index_col=0)
df
```

```python
#finding how much people bought car in every year using groupby
df.groupby(['year'])['name'].count()
```

```python
#how to change groupby object to dataframe
#we can use reset_index() fuction to convert groupby object to dataframe
d=df.groupby(['year'])['name'].count()
dd=d.reset_index()
dd.columns=['year','name']#here we giving index position
dd
```

```python
#not only groupby object,we can use reset_index() for all other object like count_value()
#here we finding total count of each cylinders and changing to dataframe
dd=df['cylinder'].value_counts()
dd_to_dataframe=dd.reset_index()
dd_to_dataframe
```
