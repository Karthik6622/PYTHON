```python
#creating dataset with columns as we checked before
import pandas as pd
li1=[item for item in range(0,64)]
df=pd.DataFrame(columns=li1)
df
```

```python
#here we extracing the image as pixel by opencv
import cv2
import numpy as np
from sklearn.preprocessing import StandardScaler

fruits=['APPLE','BANANA','MANGO','GRAPHES']
labels=[]
column_names=df.columns
for i in fruits:
    for j in range(1,11):
        image = cv2.imread(r"C:\Users\Admin\PYTHON\Machine_Learning\Support vector Machine algorithm\FRUIT_IMAGES\{}\Image_{}.jpg".format(i,j))

        # Convert the image to grayscale
        gray_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

        # Threshold the image to create a binary image
        _, binary_img = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)

        # Find contours in the binary image
        contours, _ = cv2.findContours(binary_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Sort the contours by area and get the largest contour (assuming it's the number 3)
        contours = sorted(contours,key=cv2.contourArea, reverse=True)
        largest_contour = contours[0]

        # Get the bounding box of the largest contour
        x, y, w, h = cv2.boundingRect(largest_contour)

        # Crop the region of interest (ROI) containing the number 3
        number_three_roi = gray_img[y:y + h, x:x + w]

        # Resize the ROI to 8x8 pixels
        resized_roi = cv2.resize(number_three_roi, (8, 8))

        # Flatten the 2D array to a 1D array
        arr = resized_roi.flatten()

        # Reshape the 1D array to a 2D array with a single sample
        #arr_2d = arr.reshape(1, -1)
        label=i
        d={}
        index=len(df)
        for ii,jj in zip(column_names,arr):
            d.update({ii:jj})
        df.loc[index]=d
        labels.append(label)
```

```python
#seggregate the value to train and test
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
#applying the SVM algorithm
from sklearn import svm
sv1=svm.SVC(kernel="rbf")
sv1.fit(X_train,Y_train)
```
