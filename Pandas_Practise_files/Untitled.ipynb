# getting the Cheif minister datas from wikipedia

```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://en.wikipedia.org/wiki/List_of_chief_ministers_of_Tamil_Nadu")
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
#f=soup.find(_class=)
k=soup.find(class_="wikitable sortable jquery-tablesorter")
s=str(k)
content = BeautifulSoup(s,'lxml').get_text()
print(content)
```
