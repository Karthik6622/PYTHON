# circular Linkedlist

```python
#circular singly linkedlist
class Node:
    def __init__(self,data):
        self.data=data
        self.ref=None
class DLL():
    def __init__(self):
        self.head=None
        self.end=None
    def empty(self,data):
        if self.head is None:
            new_node=Node(data)
            new_node.ref=new_node
            self.head=new_node
            self.end=new_node 
        else:
            print("Doubly linked List is not Empty!!!")
    
    def add_begin(self,data):
        new_node=Node(data)
        
        if self.head is None:
            new_node.ref=new_node
            self.head=new_node
            self.end=new_node 
        else:
            new_node.ref=self.head
            self.head=new_node
            self.end.ref=new_node 
    def add_end(self,data):
        new_node=Node(data)
        if self.head is None:
            new_node.ref=new_node
            self.head=new_node
            self.end=new_node 
        else:
            new_node.ref=self.head
            self.end.ref=new_node 
            self.end=new_node
    def add_before(self,data,x):
        if self.head is None:
            print("Linked list is Empty!!!")
            return
        if self.head.data==x:
            new_node=Node(data)
            new_node.ref=self.head
            self.head=new_node
            self.end.ref=new_node
            return
        n=self.head
        while n is not self.end:
            if n.ref.data==x:
                break
            n=n.ref
        if n.ref.data !=x:
            print("Node not found!!!")
        else:
            new_node=Node(data)
            new_node.ref=n.ref
            n.ref=new_node
        
    def add_After(self,data,x):
        if self.head is None:
            print("Linked list is Empty!!!")
            return
        n=self.head
        while n is not self.end:
            if n.data==x:
                break
            n=n.ref
        if n.data !=x:
            print("Node not found!!!")
            return
        if n.data == self.end.data:
            new_node=Node(data)
            new_node.ref=n.ref
            n.ref=new_node
            self.end=new_node
        else:
            new_node=Node(data)
            new_node.ref=n.ref
            n.ref=new_node
    def d_b(self):
        if self.head is None:
            print("Linked list is Empty!!!")
        else:
            if self.head != self.end:
                self.head=self.head.ref
                self.end.ref=self.head 
            else:
                self.head=self.end=None
    def d_end(self):
        if self.head is None:
            print("Linked list is Empty!!!")
            return
        elif self.head == self.end:
            self.head=self.end=None
        else:
            n=self.head
            while n.ref != self.end:
                n=n.ref
            self.end=n
            self.end.ref=self.head
    def d_m(self,x):
        if self.head is None:
            print("Linked list is Empty!!!")
            return
        elif self.head.data==x:
            self.d_b()
        elif self.end.data==x:
            self.d_end()
        else:
            n=self.head
            while n.ref is not self.end:
                if n.ref.data==x:
                    break
                else:
                    n=n.ref
            if n.ref!=x:
                print("Node not found!!!")
            else:
                n.ref=n.ref.ref
    def traversal(self):
        if self.head is None:
            print("CLL is Empty!!!")
        else:
            n=self.head
            while n.ref:
                print(n.data,'-->',end="")
                n=n.ref
                if n==self.head:
                    break
           

sa=DLL()

sa.empty(10)
sa.add_end(35)
sa.add_end(350)

sa.traversal()
```

```python
#circular doubly linked list
class Node:
    def __init__(self,data):
        self.data=data
        self.nref=None
        self.pref=None
class DLL():
    def __init__(self):
        self.head=None
        self.end=None
    #forward traversal
    def forward_traversal(self):
        if self.head is None:
            print("Doubly Linked list is Empty!!!")
        else:
            n=self.head
            while n:
                print(n.data,"-->",end="")
                n=n.nref
                if n==self.head:
                    break
    #backward traversal
    def backward_traversal(self):
        print()
        if self.head is None:
            print("Doubly Linked list is Empty!!!")
        else:
            n=self.end
            while n:
                print(n.data,"-->",end="")
                n=n.pref
                if n==self.end:
                    break
    #Insertion when linked list is empty
    def insert_empty(self,data):
        if self.head is None:
            new_node=Node(data)
            new_node.nref=new_node
            new_node.pref=new_node
            self.head=new_node
            self.end=new_node
        else:
            print("Doubly linked List is not Empty!!!")
    #adding node at the Begining
    def add_begin(self,data):
        new_node=Node(data)
        if self.head is None:
            new_node.nref=new_node
            new_node.pref=new_node
            self.head=new_node
            self.end=new_node
        else:
            new_node.nref=self.head
            new_node.pref=self.end
            self.end.nref=new_node
            self.head.pref=new_node
            self.head=new_node
            
    #adding node at the end
    def add_end(self,data):
        new_node=Node(data)
        if self.head is None:
            new_node.nref=new_node
            new_node.pref=new_node
            self.head=new_node
            self.end=new_node
        else:
            n=self.head
            while n is not self.end:
                n=n.nref
            new_node.nref=self.head
            new_node.pref=n
            n.nref=new_node
            self.end=new_node
            self.head.pref=new_node
    #adding node after the we mentioned node
    def add_after(self,data,x):
        if self.head is None:
            print("doubly linked list is Empty!!!")
        else:
            n=self.head
            while n is not self.end:
                if n.data==x:
                    break
                n=n.nref
            if n.data!=x:
                print("Mentiond node was not found!!!")
            else:
                new_node=Node(data)
                new_node.nref=n.nref
                new_node.pref=n
                 #checking given node not last value
                if n.nref is not self.head:
                    n.nref.pref=new_node
                    n.nref=new_node
                else:
                    n.nref=new_node
                    self.end=new_node
                    self.head.pref=new_node
    #adding node before the we mentioned node
    def add_before(self,data,x):
        if self.head is None:
            print("doubly linked list is Empty!!!")
        else:
            n=self.head
            while n is not self.end:
                if n.data==x:
                    break
                n=n.nref
            if n.data!=x:
                print("Mentiond node was not found!!!")
            else:
                new_node=Node(data)
                new_node.nref=n
                new_node.pref=n.pref
                #checking given node not first value
                if n.pref is not self.end:
                    n.pref.nref=new_node
                    n.pref=new_node
                else:
                    self.head=new_node
                    self.end.nref=new_node
                    n.pref=new_node
                               
      #deleting node at the begining
    def delete_begin(self):
        if self.head is None:
            print("doubly linked list is Empty!!!")
        elif self.head==self.end:
            self.head=self.end=None
            print("DLL Going to empty after Deleting this one!!!1")
        else:
            self.head=self.head.nref
            self.head.pref=self.end
            self.end.nref=self.head
    #deleting node at the end
    def delete_end(self):
        if self.head is None:
            print("doubly linked list is Empty!!!")
        elif self.head == self.end:
            self.head=self.end=None
            print("DLL Going to empty after Deleting this one!!!1")
        else:
            n=self.head
            while n is not self.end:
                n=n.nref
            self.end=n.pref
            n.pref.nref=self.head
            self.head.pref=self.end
    #deleting node by the values
    def delete_value(self,x):
        # checking DLL is Empty 
        if self.head is None:
            print("doubly linked list is Empty!!!")
        #checking only one node present
        elif self.head==self.end:
            if self.head.data==x:
                self.head=self.end=None   
        #checking user mentioned node is first node 
        elif self.head.data==x:
            self.head=self.head.nref
            self.head.pref=self.end
            self.end.nref=self.head
            
        else:
            n=self.head
            while n is not self.end:
                if n.data==x:
                    break
                n=n.nref
            #checking user mentioned node is middle node
            if n is not self.end:
                n.nref.pref=n.pref
                n.pref.nref=n.nref
            else:
                #checking user mentioned node is last node
                if x==n.data:
                    self.end=n.pref
                    n.pref.nref=self.head
                    self.head.pref=n.pref
                    
                else:
                    print("Mentiond node was not found!!!")
                   
                    
d=DLL()
d.add_end(2301)
d.add_before(70,2301)
d.add_before(700,2301)
d.add_before(700,2301)
d.delete_value(2301)
d.delete_value(70)
d.delete_value(701)
d.forward_traversal()
d.backward_traversal()
```
