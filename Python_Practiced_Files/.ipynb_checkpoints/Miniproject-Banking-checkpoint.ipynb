# Miniproject-Banking

```python
dic={}
```

```python

import random
k=True
while(k==True):
    print('*'*50)
    print("WELCOME TO TIMES BANK")
    print('*'*50)
    print("1.OPEN A NEW ACCOUNT")
    print("2.WITHDRAW MONEY")
    print("3.DEPOSIT MONEY")
    print("4.CHECK CUSTOMER & BALANCE")
    print("5.EXIT/QUIT")
    print("6.ADMIN")
    print('*'*50)
    inpu=int(input("ENTER YOUR NUMBER FOR OPERATION:"))
    if(inpu==1):
        USER_NAME=input("PLEASE ENTER YOUR NAME:")
        if((USER_NAME>='a' and USER_NAME<='z')or(USER_NAME>='A' and USER_NAME<='Z')):
            BIN=int(input("PLEASE ENTER YOUR 4 digit bin:"))
            RETYPE_BIN=int(input("PLEASE ENTER YOUR 4 digit bin AGAIN:"))
            if (BIN==RETYPE_BIN) and ((len(str(BIN))>=4) and (len(str(RETYPE_BIN))>=4)):
                DEPOSIT_AMOUNT=int(input("PLEASE ENTER YOUR INITIAL DEPOSIT AMOUNT minimum100 to maximum10000:"))
                if(DEPOSIT_AMOUNT>=100 and DEPOSIT_AMOUNT<=10000):
                    print("Account created")
                    ACCOUNT_NUMBER=random.randint(11111,99999)
                    print("YOUR ACCOUNT NUMBER IS:",ACCOUNT_NUMBER)
                    s=str(USER_NAME)
                    s1=str(BIN)
                    s2=str(DEPOSIT_AMOUNT)
                    dic[ACCOUNT_NUMBER]='{},{},{}'.format(s,s1,s2)
                    
                else:
                    print("PLEASE ENTER THE MINIMUM AMOUNT 100 and MAXIMUM 10000")
                    print("PLEASE ENTER CORRECT INFORMATION AND TRY AGAIN")
            else:
                print("Please ENTER 4 digit bin and both should same")
                print("PLEASE ENTER CORRECT INFORMATION AND TRY AGAIN")
        else:
            print("Please enter the valid name")
            print("PLEASE ENTER CORRECT INFORMATION AND TRY AGAIN")
    elif(inpu==2):
        withdraw=int(input("Enter the Withdrow Amount(100/200/500/2000):"))
        if(withdraw==100 or withdraw==200 or withdraw==500 or withdraw==2000):
            accont_vali=int(input("Enter the Account Number:"))
            bin1=int(input("Enter the BIN Number:"))
            li=list(dic.keys())
            k1=len(li)
            j=0
            s1=''
            for i in li:
                if(accont_vali==i):
                    s1=dic[i]
                    s2=list(s1.split(","))
                    if(s2[1]==str(bin1)):
                        if(int(s2[2])>=100 and int(s2[2])>withdraw):
                            balance_amount=int(s2[2])-withdraw
                            print("money has been withdrawn")
                            print("BALANCE AMOUNT:",balance_amount)
                            s=s2[0]
                            s1=s2[1]
                            s2=str(balance_amount)
                            dic[accont_vali]='{},{},{}'.format(s,s1,s2)
                        else:
                            print("Money not there!!!please check your account balance")
                    else:
                        print("PLEASE ENTER CORRECT BIN NUMBER")
                else:
                    j+=1
            if(j==k1):
                print("PLEASE ENTER CORRECT ACCOUNT NUMBER")
        else:
            print("please enter amount as we mentioned")
        
    elif(inpu==3):
        deposit=int(input("Enter the Deposit Amount:"))
        if(deposit<=10000):
            accont_vali=int(input("Enter the Account Number:"))
            bin1=int(input("Enter the BIN Number:"))
            li=list(dic.keys())
            k1=len(li)
            j=0
            s1=''
            for i in li:
                if(accont_vali==i):
                    s1=dic[i]
                    s2=list(s1.split(","))
                    if(s2[1]==str(bin1)):
                        if (int(s2[2])+deposit<100000):
                            total_amount=int(s2[2])+deposit
                            print("money has been deposited")
                            print("TOTAL AMOUNT:",total_amount)
                            s=s2[0]
                            s1=s2[1]
                            s2=str(total_amount)
                            dic[accont_vali]='{},{},{}'.format(s,s1,s2)
                        else:
                            print("Please keep your account amount should be less than 100000")
                    else:
                        print("PLEASE ENTER CORRECT BIN NUMBER")
                else:
                    j+=1
            if(j==k1):
                print("PLEASE ENTER CORRECT ACCOUNT NUMBER")
        else:
            print("Deposit amount should be 10000 or lessthan 10000")
    elif(inpu==4):
        accont_vali=int(input("Enter the Account Number:"))
        bin1=int(input("Enter the BIN Number:"))
        li=list(dic.keys())
        k1=len(li)
        j=0
        s1=''
        for i in li:
            if(accont_vali==i):
                s1=dic[i]
                s2=list(s1.split(","))
                if(s2[1]==str(bin1)):
                    print('NAME:',s2[0])
                    print('BIN:',s2[1])
                    print('AMOUNT:',s2[2])  
                else:
                    print("PLEASE ENTER CORRECT BIN NUMBER")
            else:
                j+=1
        if(j==k1):
            print("PLEASE ENTER CORRECT ACCOUNT NUMBER")
        
    elif(inpu==5):
        k=False
    elif(inpu==6):
        username=input("Enter the Admin username:")
        password=int(input("Enter the Admin password:"))
        if(username=='admin' and password==123456):
            jj=list(dic.keys())
            for l in jj:
                ssss=dic[l]
                ssss1=list(ssss.split(","))
                print("NAME:",ssss1[0])
                print("ACCOUNT NUMBER:",l)
                print("BIN:",ssss1[1])
                print("AMOUNT:",ssss1[2])
                print("-"*100)
                
        else:
            print("PLEASE ENTER CORRECT USERNAME and PASSWORD")
    else:
        print("PLEASE SELECT (1/2/3/4/5/6) for your operation")
```
