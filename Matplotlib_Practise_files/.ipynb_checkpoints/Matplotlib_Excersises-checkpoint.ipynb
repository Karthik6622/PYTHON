```python
# Create an empty subplot with 2 rows and 2 columns (4 subplots total)
fig,s=plt.subplots(2,2)
```

```python
#instead of empty plot created easy plots
# Create an empty subplot with 2 rows and 2 columns (4 subplots total)
plt.subplot(2,2,1)
plt.plot([90,4,3,2])
plt.subplot(2,2,2)
plt.plot([90,40,3,2])
plt.subplot(2,2,3)
plt.plot([0,4,3,2])
plt.subplot(2,2,4)
plt.plot([90,44,3,26])
```

```python
#using subplots
# Create the same plot as above with 2 rows and 2 columns and figsize of (10, 5)
fig,s=plt.subplots(2,2,figsize=(10,5))
# Plot X versus X/2 on the top left axes
x=np.array([45,32,344,1245])
s[0,0].plot(x,x/2,color='b',marker='p',ms=10,mfc='orange')
# Plot a scatter plot of 10 random numbers on each axis on the top right subplot
random_numbers1=np.random.randint(1,10,10)
random_numbers2=np.random.randint(1,10,10)
s[0,1].scatter(random_numbers1,random_numbers2)
# Plot a bar graph of the favourite food keys and values on the bottom left subplot
dic={'briyani':200,'Idly':30,'pongal':40,'meals':80,'Porrata':50}
s[1,0].bar(dic.keys(),dic.values())
s[1,0].set_xlabel("FOOD")
s[1,0].set_ylabel("PRICE")
s[1,0].set_title("FAVORITE FOOD")
# Plot a histogram of 1000 random normally distributed numbers on the bottom right subplot
ndn=np.random.randn(1000)
s[1,1].hist(ndn,color='yellow',edgecolor='black')
```

Notice how the subplot has multiple figures. Now let's add data to each axes.

```python
#using figure
# Create the same plot as above with 2 rows and 2 columns and figsize of (10, 5)
fig=plt.figure(figsize=(10,5))
axis1=fig.add_subplot(2,2,1)
axis2=fig.add_subplot(2,2,2)
axis3=fig.add_subplot(2,2,3)
axis4=fig.add_subplot(2,2,4)
# Plot X versus X/2 on the top left axes
x=np.array([45,32,344,1245])
axis1.plot(x,x/2,color='b',marker='p',ms=10,mfc='orange')
# Plot a scatter plot of 10 random numbers on each axis on the top right subplot
random_numbers1=np.random.randint(1,10,10)
random_numbers2=np.random.randint(1,10,10)
axis2.scatter(random_numbers1,random_numbers2)
# Plot a bar graph of the favourite food keys and values on the bottom left subplot
dic={'briyani':200,'Idly':30,'pongal':40,'meals':80,'Porrata':50}
axis3.bar(dic.keys(),dic.values())
axis3.set_xlabel("FOOD")
axis3.set_ylabel("PRICE")
axis3.set_title("FAVORITE FOOD")
# Plot a histogram of 1000 random normally distributed numbers on the bottom right subplot
ndn=np.random.randn(1000)
axis4.hist(ndn,color='yellow',edgecolor='black')
```

Why doesn't it work?

Hint: It's not numeric data.

In the process of turning it to numeric data, let's create another column which adds the total amount of sales and another one which shows what date the car was sold.

Hint: To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
#converted Price column to String
df['Price']=df['Price'].astype(str)
#removing $ symbol and changed to numbers
df['Price']=df['Price'].str.replace('[$-,]','',regex=True).astype(float)
#'[$-,]'=this line of code uses regular expressions to remove any occurrences of '$' and ',' characters from the 'Price' column.
df['Price'].plot(kind="bar")
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
df['Total Sales']=df['Price'].cumsum()
# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
   #finding today date
import datetime
today=datetime.date.today()
   #finding length of dataframe
len_of_dataframe=len(df)
   #adding 'Sale Date' column using pd.date_range() function
dd=pd.date_range(start=today,periods=len_of_dataframe)
df['Sale Date']=dd
# View the car_sales DataFrame
df
```

Now we've got a numeric column (Total Sales) and a dates column (Sale Date), let's visualize them.

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
  #plt.plot(df['Total Sales'],df['Sale Date'])
      #  or
df.plot(x='Total Sales',y='Sale Date')
```

```python
# Convert the 'Price' column to the integers
df['Price']=df['Price'].astype(int)
# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
df.plot(kind='scatter',x='Odometer (KM)',y='Price',color='y')
```

Now let's try comparing two variables versus the target variable.

More specifially we'll see how age and cholesterol combined effect the target in patients over 50 years old.

For this next challenge, we're going to be replicating the following plot:

```python
# Replicate the above plot in whichever way you see fit
# Note: The method below is only one way of doing it, yours might be
# slightly different

# Create DataFrame with patients over 50 years old
dataf_above_50=heart_disease[heart_disease['age']>50]
#dataf_above_50
# Create the plot
fig,ax=plt.subplots()
# Plot the data
ax.scatter(dataf_above_50['age'],dataf_above_50['chol'],edgecolor='black',c=dataf_above_50['target'])
#c=dataf_above_50['target']=target column has two values(1,0),here it give 1 means one color and 0 means one color
# Customize the plot
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol Level')
ax.set_title('Age vs. Cholesterol for Patients Over 50')
# Add a meanline
chol_mean=dataf_above_50['chol'].mean()#finding  average value or mean value of chol column
age_mean=dataf_above_50['age'].mean()#finding average value or mean value of age column
ax.axvline(age_mean,color='red',linestyle="--",label="mean age")#axvline is used to draw the line vertically inside the  plot
ax.axhline(chol_mean,color='red',linestyle="--",label="mean chol")#axhline is used to draw the line horizontaly inside the  plot
#we had drwn the line in the mean value of chol and mean value of age
ax.legend()
```

```python
# Reproduce the same figure as above with the "seaborn-whitegrid" style
# Create DataFrame with patients over 50 years old
dataf_above_50=heart_disease[heart_disease['age']>50]
#dataf_above_50
# Create the plot
fig,ax=plt.subplots(figsize=(10,6))
# Plot the data

ax.scatter(dataf_above_50['age'],dataf_above_50['chol'],edgecolor='black',c=dataf_above_50['target'])
#c=dataf_above_50['target']=target column has two values(1,0),here it give 1 means one color and 0 means one color
# Customize the plot

ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol Level')
ax.set_title('Age vs. Cholesterol for Patients Over 50')
# Add a meanline
hol_mean=dataf_above_50['chol'].mean()#finding  average value or mean value of chol column
age_mean=dataf_above_50['age'].mean()#finding average value or mean value of age column
ax.axvline(age_mean,color='red',linestyle="--",label="mean age")#axvline is used to draw the line vertically inside the  plot
ax.axhline(chol_mean,color='red',linestyle="--",label="mean chol")#axhline is used to draw the line horizontaly inside the  plot
#we had drwn the line in the mean value of chol and mean value of age
ax.legend()
```

Wonderful, you've changed the style of the plots and the figure is looking different but the dots aren't a very good colour.

Let's change the cmap parameter of scatter() as well as the color parameter of axhline() to fix it.

Completing this step correctly should result in a figure which looks like this: 

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"
fig,ax=plt.subplots(figsize=(10,6))
# Plot the data

ax.scatter(dataf_above_50['age'],dataf_above_50['chol'],edgecolor='black',c=dataf_above_50['target'],cmap="winter")
#cmap(colormap)=it is used to provide color for plot
#c=dataf_above_50['target']=target column has two values(1,0),here it give 1 means one color and 0 means one color
# Customize the plot

ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol Level')
ax.set_title('Age vs. Cholesterol for Patients Over 50')
# Add a meanline
hol_mean=dataf_above_50['chol'].mean()#finding  average value or mean value of chol column
age_mean=dataf_above_50['age'].mean()#finding average value or mean value of age column
ax.axvline(age_mean,color='red',linestyle="--",label="mean age")#axvline is used to draw the line vertically inside the  plot
ax.axhline(chol_mean,color='red',linestyle="--",label="mean chol")#axhline is used to draw the line horizontaly inside the  plot
#we had drwn the line in the mean value of chol and mean value of age
ax.legend()
```
