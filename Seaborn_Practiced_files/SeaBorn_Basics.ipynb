# seaborn
Seaborn offers lot of functionality which makes it effective for many data visualization tasks. These functionalities are listed below:-

• Seaborn provides a dataset-oriented API to examine relationships between variables.

• It provides functions to fit and visualize linear regression models for different types of independent and dependent variables.

• Seaborn provide functions for visualizing univariate and bivariate distributions and for comparing them between subsets of data.

• It provides plotting functions for using categorical variables to show observations or aggregate statistics.

• It helps us to visualize matrices of data and use clustering algorithms to discover structure in those matrices.

• Seaborn provides a plotting function to plot statistical time series data. The function provides flexible estimation and representation of uncertainty around the estimate.

• It provide tools for choosing styles, colour palettes, palette widgets and utility functions. These tools help us to make beautiful plots that reveal patterns in the data.

• It provides several built-in themes for producing stylish looking Matplotlib graphics.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#importing seaborn 
import seaborn as sns
```

```python
#seaborn has some inbuilt dataset
#we can use those dataset while we need
#how to see those dataset
data=sns.get_dataset_names()
print(data)
```

```python
#these are the dataset we can use for our use
data
```

```python
#using load_dataset class we can use tips dataset
tips=sns.load_dataset('tips')
tips
```

```python
#if we want,titanic dataset means use below code
titanic=sns.load_dataset('titanic')
titanic
```

```python
#barplot
#it is used to make the data's bar shape visualization
sns.barplot(x='day',y='total_bill',data=tips)#tips=dataset
#here x,y we given tips dataset column names
plt.show()
```

```python
#violinplot
#it is used to show the data as violin shape
sns.violinplot(x=tips['total_bill'])
```

```python
sns.violinplot(x=tips['day'],y=tips['total_bill'],data=tips)
```

```python
#countplot-it is used to count the data(example: in dataset how much male and female are there,how much people smoke and not smoke)
sns.countplot(x='smoker',data=tips)
#here we displayed how much people smoke and not smoke
plt.show()
```

```python
#how to find male and female count using countplot
sns.countplot(x='sex',data=tips)
```

```python
#how to find how many male and female smoking in tips dataset
sns.countplot(x='smoker',data=tips,hue='sex')
```

```python
#barplot
sns.barplot(x='total_bill',y='day',data=tips)
```

```python
#barplot
#how to use barplot estimator parameter
sns.barplot(x='total_bill',y='day',data=tips,estimator=np.max)
#here using estimator we finding max value of total_bill column(we can't give non-numeric column)
#If you don't specify the estimator parameter, Seaborn will use the mean as the default estimator
#we can use np.min,np.count,np.sum,np.median
```

```python
#set_style()-using this function we can set background color and style for plot
sns.set_style('dark')#dark-background dark theme
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.set_style('darkgrid')#dark-background dark theme
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.min)
#Certainly! The seaborn.set_style() function in the Seaborn data visualization library is used to set the aesthetic style of the plots you create. It allows you to control various attributes of the plot's appearance. Here are some of the attributes you can set using seaborn.set_style()
'''
darkgrid': This style uses a dark background grid, which is suitable for emphasizing the data points in your plot.
'whitegrid': Similar to 'darkgrid', but with a white background grid.
'dark': This style removes the background grid, resulting in a plain, dark background.
'white': Similar to 'dark', but with a white background.
'ticks': This style shows only the ticks on the axes, creating a minimalistic look.
'''
```

```python
#how to set custom style
custom_style = {
    'axes.facecolor': 'red',  # Background color of the plot area
    'grid.color': 'white',         # Color of the grid lines
}
sns.set_style('whitegrid', rc=custom_style)#rc-resource configuration attribute
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.min)
```

# Seaborn Histogram
Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

We can use the same distplot() function to plot a Histogram

```python
#histogram-it's represent a data distribution like barplot
#it's same like a distplot() histogram
sns.set_style('white')
sns.histplot(x='total_bill',data=tips)#we can't give both x and y.anyone only we can give
```

```python
#to draw histogram plot in seaborn we use distplot
#distplot-this only availble in lower vesrions.in higher vesrion seaborn they replaced displot or histplot
sns.distplot(tips['total_bill'])#we can't give both x and y.anyone only we can give
```

```python
#kde-#kernel density estimates
#it will display density of total_bill column
sns.distplot(tips['total_bill'],kde=False)#kde-is just that line ,#we can't give both x and y.anyone only we can give
```

```python
sns.distplot(tips['total_bill'],hist=False)#we can't give both x and y.anyone only we can give
#hist-histogram.if we make histogram false,only kde will display
#histogram display total_bill column count
```

```python
#displot-(distribution plot)it's a advance version of distplot but like a same
sns.set_style('whitegrid')
sns.displot(x='total_bill',data=tips,kde=True)#we can't give both x and y.anyone only we can give
```

```python
sns.set_style('white')
sns.displot(x='day',data=tips)#defaulty it's only give hist
#kde=density of given column
#hist=count of given column
```

```python
#we cant use hist=False in displot and histplot
sns.displot(x='day',data=tips,hist=False,kde=True)
```

```python
#kdeplot if we want only kde means we can use kdeplot
sns.kdeplot(tips['total_bill'])
plt.show()
```

#jointplot() - creates a multi-panel figure that projects the bivariate relationships between two variables and also the univariate distribution of each variable on separate axes.

```python
#jointplot-it is used to create scatterplot between given two colums and histogram for both column
#bivariate-scatterplot
#univariate-histogram
sns.jointplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
#scatterplot-Scatterplots are useful for visualizing the relationship between two continuous variables
sns.scatterplot(x='tip',y='total_bill',data=tips)
```

```python
#pairplot- function is used to create a grid of scatterplots and histograms for multiple variables in a dataset. It's a convenient way to visualize the pairwise relationships between numerical variables in a dataset
sns.pairplot(data=tips)
#same column means histogram
#different coilmn means scatter
```

```python
#stripplot- it's like a scatterplot but display dots clearly same line of x variable
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
#how to chnge stripplot colors
colors=['red','yellow','green','pink']
sns.stripplot(x='day',y='total_bill',data=tips,palette=colors)#using palette we can chane colors
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
#This helps avoid having multiple points with the same value stacked on top of each other
#to prevent overlapping points and provide a better visualization of the distribution of data.
```

```python
#boxplot-
#first line is minimum value of total_bill column
#box starting line is 25% value of total_bill column
#box center line is 50% value of total_bill column
#box last line is 75% value of total_bill column
#last end diamond is Max value of total_bill column
sns.boxplot(x=tips['total_bill'])
```

lmplot() - plot data and regression model fits across a facetGrid.

```python
#lmplot-(linear model plot)it is used to draw scatter and linear regression line for we given data's
#linear regression line show whteher value increased or not()
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
#how to draw only linear regression line
x=np.array([1,2,3,4,5])
y=np.array([5,4,3,2,1])
data=pd.DataFrame({'x':x,'y':y})
sns.lmplot(x='x',y='y',data=data,scatter=False)
plt.show()
```

```python
#if we want olny scatter use below code
sns.lmplot(x='x',y='y',data=data,fit_reg=False)
```

```python
#residplot
#In Seaborn, residplot is a function that creates a residual plot.
#A residual plot is used to visualize the residuals or errors in a regression analysis. 
#Residuals represent the difference between the observed (actual) values and the predicted values from a regression model. 
sns.residplot(x='total_bill',y='tip',data=tips)
#it's shows all total bill value as well as predicted values
```

```python
x=np.array([3,6,9,1])
y=np.array([30,60,90,10])
data=pd.DataFrame({'x':x,'y':y})
sns.residplot(x='x',y='y',data=data)
```
