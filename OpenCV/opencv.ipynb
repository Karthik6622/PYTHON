```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
img=cv.imread('ronald1o.jfif')
assert img is not None,"file not found karthik"
#The assert statement is used to check if img is not None. If img is None, the specified error message, "File not found" will be displayed,
cv.imshow("Image",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#applying 2d filter to the image
img=cv.imread('ronaldo.jfif')
#ones arry will change the image quality as we given element of array
kernel=np.ones((5,10),np.float32)/50
dst=cv.filter2D(img,-1,kernel)
#-1 as the second argument specifies the depth of the output image. In this case, -1 means the depth will be the same as the input image.
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("original image")
plt.xticks([])#it's remove x values(like 200,400)
plt.yticks([])#it's remove y values(like 200,400)
plt.subplot(1,2,2)
plt.imshow(dst)
plt.title("filtered image")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#how to create blur on the image
#if we display image by matplotlib it will show the image like a different color because it show image in rgb format
#if we want display image as same original we have to convert image rgb to bgr format
img=cv.imread("ronaldo.jfif")
#covert image in rgb format to bgr format
converted=cv.cvtColor(img,cv.COLOR_RGB2BGR)
blur=cv.blur(converted,(5,5))#it's apply 5*5 matrix on image for bluring
plt.subplot(1,2,1)
plt.imshow(converted)
plt.subplot(1,2,2)
plt.imshow(blur)
```

```python
#Gussain blur-it is one type blurring method.it is used to blur the image
img=cv.imread("ronaldo.jfif")
#covert image in rgb format to bgr format
converted=cv.cvtColor(img,cv.COLOR_RGB2BGR)
blur=cv.GaussianBlur(converted,(5,5),0)#it's only ccept odd number matrix like a 5*5,7*7,9*9
'''
(5, 5) specifies the size of the Gaussian kernel or filter. In this case, it's a 5x5 kernel. The size of the kernel determines the extent of blurring. Larger kernels result in stronger blurring.

0 specifies the standard deviation in the X direction of the Gaussian kernel. A standard deviation of 0 means that OpenCV will automatically calculate it based on the kernel size. You can provide a non-zero value if you want to control the standard deviation.

The result of this operation is a blurred version of the converted image, which is stored in the variable blur. Gaussian blur is a common technique used in image processing to reduce noise and smooth an image. It works by convolving the image with a Gaussian kernel, which assigns a weighted average of nearby pixel values to each pixel in the output image, effectively smoothing out details and reducing high-frequency noise.
'''
plt.subplot(1,2,1)
plt.imshow(converted)
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(blur)
plt.title("blur image")
```

```python
#medianblur-it is one type blurring method.it is used to blur the image
img=cv.imread("ronaldo.jfif")
#covert image in rgb format to bgr format
converted=cv.cvtColor(img,cv.COLOR_RGB2BGR)
blur=cv.medianBlur(converted,5)#5-size of kernel(apply the blur by size of kernel)
plt.subplot(1,2,1)
plt.imshow(converted)
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(blur)
plt.title("blur image")
```
