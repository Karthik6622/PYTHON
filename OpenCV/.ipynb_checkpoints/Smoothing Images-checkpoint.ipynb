# Smoothing Images

Smoothing images in OpenCV is a common image processing task, often used for noise reduction or blurring. There are various methods available for image smoothing in OpenCV,

Homogeneous filter

Gaussian filter

Median filter

Bilateral filter

#Homogeneous filter

The Homogeneous filter, also known as the Box filter, is a simple and basic type of filter used in image processing for smoothing or blurring an image. It's essentially a square kernel with equal weights applied to all pixels within the kernel. The result of applying a Homogeneous filter is a blurred version of the image.

```python
#read the image form system using opencv
import cv2
img=cv2.imread("ronaldo.jfif")
kernel=(15,5)
#You can adjust the kernel_size to control the amount of smoothing or blurring applied to the image. Larger kernel sizes result in more significant smoothing.
homogenious_filter=cv2.boxFilter(img,-1,kernel)
#The -1 in the function call indicates that the output image will have the same depth as the input image.
cv2.imshow("Normal Image",img)
cv2.imshow("Homogeneous Filter",homogenious_filter)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#display homogenious filtered images using matlotlib
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Normal image")
plt.subplot(1,2,2)
plt.imshow(homogenious_filter)
plt.title("homogenious filtered image")
```

Gaussian Filter

The Gaussian filter is a popular and effective filter used for smoothing or blurring images in image processing. It uses a Gaussian kernel, which has a bell-shaped distribution of weights, giving more importance to pixels closer to the center of the kernel. This results in a smoother and more natural-looking blur compared to the Homogeneous (Box) filter.

```python
#read image
img=cv2.imread("ronaldo.jfif")
#converting rgb format to bgr format because of we display image by matplolib
img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
kernel_area=(15,5)# Define the size of the kernel (should be an odd number)
sigma_x=0
sigma_y=0
gaussian_filter=cv2.GaussianBlur(img,kernel_area,sigma_x,sigma_y)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Normal image")
plt.subplot(1,2,2)
plt.imshow(gaussian_filter)
plt.title("gaussian filtered image")
```

Median filter

The Median filter is a nonlinear filter commonly used in image processing for noise reduction and removing salt-and-pepper noise, which are black and white specks that can corrupt an image. The Median filter replaces each pixel's value with the median value within a local neighborhood around that pixel.

```python
import cv2

# Load an image
image = cv2.imread('ronaldo.jfif')

# Define the size of the kernel (should be an odd number)
kernel_size = 5  # Adjust this based on your preference

# Apply the Median filter
median_filtered = cv2.medianBlur(image, kernel_size)

# Display the original and filtered images
cv2.imshow('Original Image', image)
cv2.imshow('Median Filtered Image', median_filtered)

# Wait for a key press and then close the windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Bilateral filter

The Bilateral filter is a popular image filter in computer vision and image processing. It is used for smoothing or denoising images while preserving edges and fine details.

```python
#read an image
image=cv2.imread('ronaldo.jfif')
#diameter" refers to the size of the pixel neighborhood that the filter considers when processing each pixel in the image.
diameter=15
#Define the standard deviation of the color space
sigma_color = 100  # Adjust this based on your preference
# Define the standard deviation of the coordinate space
sigma_space = 100  # Adjust this based on your preference
bilateral_image=cv2.bilateralFilter(image,diameter,sigma_color,sigma_space)
cv2.imshow('normalimage',image)
cv2.imshow('Bilateral filter image',bilateral_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
