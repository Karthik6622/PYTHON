# display the image by matplotlib

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#read image and opening image using by matplotlib
img=cv2.imread("ronaldo.jfif")
cv2.imshow("WINDOWIMAGE",img)#here image read by BGR format
cv2.waitKey(4000)
cv2.destroyAllWindows()
plt.imshow(img)#in matplotlib image read by RGB format that's why it was showing as color different
plt.show()
```

```python
#if we want to display image as same as cv2 format in matplotlib,we have to convert image in rgb format to bgr format
converting_bgr=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(converting_bgr)
```

```python
#how to remove X axis values and Y axis values while displaying th eimage in matplotlib
plt.imshow(converting_bgr)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#how to display 6 images in one window using matplot lib
#read the 6 images
img1=cv2.imread("ronaldo.jfif")
img2=cv2.imread("image.jpg")
img3=cv2.imread("color_circle.jfif")
img4=cv2.imread("output.jpg")
img5=cv2.imread("karthik.png")
img6=cv2.imread("grayscale.jpg")
titles=['Ronaldo','Natural','Circle','Output','Karthik','Grayscale']
images=[img1,img2,img3,img4,img5,img6]
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
