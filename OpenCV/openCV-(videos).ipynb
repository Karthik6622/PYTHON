```python
#capture live stream from default camera
import cv2
#capturing video from laptop camera
cap=cv2.VideoCapture("google_issues_video.mp4")#0-for device camera.if does not work we can give -1.if we have multiple camera our device we can specify first camera as 0 and secod camera as 1 and third camera as 2

# Check if the camera opened successfully.
if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()
'''
exit() is a Python function that immediately terminates the program. 
In this context, it's used to exit the program if the camera couldn't be opened because there's no point in continuing the program if you can't access the camera.
'''
#here we using while loop to continously capture video
while True:
    # Read a frame from the camera.
    ret,frame=cap.read()
    # Check if the frame was read successfully.
    if not ret:
        print("Error: Could not read frame.")
        break
    # Display the captured frame.
    cv2.imshow('frame',frame)
    # Press 'q' to exit the video capture loop.
    key=cv2.waitKey(0)
    if key==ord('q'):
        break#it's come out of the loop
cap.release()#it's relese whatever system resource used for videcapturin#
cv2.destroyAllWindows()
```

What is codecs?

Video codecs, short for "compression-decompression," are software or hardware algorithms used to compress and decompress video data. They play a fundamental role in digital video technology by reducing the size of video files, making them more manageable for storage, transmission, and playback. Video codecs are essential for efficient video streaming, video conferencing, video storage, and many other applications involving video content. 

what is fourcc?

FOURCC stands for "Four Character Code," and it is a sequence of four characters used to uniquely identify data formats in various multimedia applications, including video codecs. These codes are often used to represent video compression formats in AVI files, among other uses.

For example, "H.264" is a widely used video codec, and its FOURCC code is often represented as "avc1" for some implementations.

```python
#how to read video from video file and write or save to another file
#read video file from system
import cv2
cap=cv2.VideoCapture("google_issues_video.mp4")
#write video to system with specified name
fourcc=cv2.VideoWriter_fourcc(*'XVID')#used to reprent a video compression formats
wri=cv2.VideoWriter('videowrite.avi',fourcc,20.0,(640,480))
#'videowrite.mp4'-name of the video saving file
#fourcc-represent video compression formats 
#20.0-per second how many frame going to save
#(468,834)-what frame width and column going to save 
while (cap.isOpened()):
    check,frame=cap.read()
    #check variable give output as true or false.it check whether frame captured or not
    if check==True:
        #writing or saving every frame to newfile
        wri.write(frame)  
        gray=cv2.cvtColor(frame,cv2.COLOR_RGB2GRAY)
        cv2.imshow("VIDEOSHOW",gray)
        if cv2.waitKey(1)==ord('q'):
            break
    else:
        break
cap.release()
wri.release()#relese all the resources which have been used for video writing
cv2.destroyAllWindows()
```

```python
#how to display video height and width on while video running and save changed video to system with specify name
cap=cv2.VideoCapture("google_issues_video.mp4")
fourcc=cv2.VideoWriter_fourcc(*'DIVX')
wri=cv2.VideoWriter("Changedvideo.mp4",fourcc,10,(400,400))
while(cap.isOpened()):
    check,frame=cap.read()
    if check == True:
        #here we getting capture video default width and height
        text='Width:'+str(cap.get(3))+' '+'Height:'+str(cap.get(4))
        font=cv2.FONT_HERSHEY_COMPLEX
        #put text on playing video
        frame=cv2.putText(frame,text,(20,50),font,1,(0,0,255),5)
        #frame-video image frame
        #txt-what text going to display on video
        #(20,50)-coordinates(starting position of text)
        #font-font of the text
        #1-size of the textfont
        #(0,0,255)-color of the text
        #5-thickness of the text
        wri.write(frame)
        cv2.imshow("Edited Video",frame)
        if cv2.waitKey(1)==ord('q'):
            break
    else:
        break
cap.release()
wri.release()
cv2.destroyAllWindows()
```

```python
#how to display current date and time on while video running 
import datetime
cap=cv2.VideoCapture("google_issues_video.mp4")
while(cap.isOpened()):
    check,frame=cap.read()
    if check == True:
        #get the current time
        date=str(datetime.datetime.now())
        font=cv2.FONT_HERSHEY_COMPLEX
        #put text on playing video
        frame=cv2.putText(frame,date,(20,50),font,1,(0,0,255),5)
        #frame-video image frame
        #txt-what text going to display on video
        #(20,50)-coordinates(starting position of text)
        #font-font of the text
        #1-size of the textfont
        #(0,0,255)-color of the text
        #5-thickness of the text
        wri.write(frame)
        cv2.imshow("Edited Video",frame)
        if cv2.waitKey(1)==ord('q'):
            break
    else:
        break
cap.release()
wri.release()
cv2.destroyAllWindows()
```
