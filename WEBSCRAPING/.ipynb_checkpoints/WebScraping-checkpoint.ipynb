```python
#if we wnat display webpage contents clearly,we can use bs4 module
from bs4 import BeautifulSoup
import requests
# get URL
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
# scrape webpage
soup=BeautifulSoup(page.content,'html.parser')#used to parse the html contents from webpage
#display scraped data
print(soup.prettify())
```

```python
#display whole html from the webpage as a list format
list(soup.children)
```

```python
#find <P> tag contents from webpage and display first <p> tag content also
# find all occurrence of p in HTML
# includes HTML tags
print(soup.find_all('p'))
print('\n')
# return only text
# does not include HTML tags
print(soup.find_all('p')[0].get_text())
```

```python
#find  id="mp-left" insie class of mp-h2 class
# create object with which tag contain id="mp-left"
object=soup.find(id="mp-left")
#sepeare mp-h2 class data from id="mp-left"
item=object.find_all(class_='mp-h2')
#display only first data
result=item[0]
print(result.prettify())
```

```python
#scrape a data's using urlopen package from another webpage(except wekipedia) and make Dataset using pandas
#urlopen-it is like a request module
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url="https://www.hubertiming.com/results/2017GPTR10K"
html=urlopen(url)
html#we got response from mentioned url
```

```python
#display type of soup
soup=BeautifulSoup(html,'lxml')#lxml-optional,we can give html.parser also
type(soup)
```

```python
#display the data
print(soup.prettify())
```

```python
#get the title
print(soup.title)
```

```python
#printout the test of webpage(no tag)
text=soup.get_text()
text
```

```python
#find all <a> tag
soup.find_all('a')
```

```python
#to get only the links in the anchor tag
all_links=soup.find_all('a')
for i in all_links:
    print(i.get('href'))
```

```python
#print the first 10 rows for sanity check
rows=soup.find_all('tr')
print(rows[:10])
```

```python
#getting <td> tag content from <tr> tags
for i in rows:
    row_td=i.find_all('td')
print(row_td)#dispay last td content
type(row_td)
```

```python
#change the <td> tag content to string
s=str(row_td)
print(s)
print(type(s))
```

```python
#change the <td> tag content to string and remove <td> tag
cleantext=BeautifulSoup(s,'lxml').get_text()
print(cleantext)
```

```python
#find out tabledata content from <tr> tag and change to dataset
import re
list_rows=[]
for row in rows:
    table_cells=row.find_all('td')
    ss=str(table_cells)
    clean=re.compile('<.*?>')
    clean2=re.sub(clean,'',ss)#sub-it's like a replace
    list_rows.append(clean2)
print(clean2)
print(type(clean2))
print(list_rows)
```

```python
df=pd.DataFrame(list_rows)
df
```

```python
#split the dataframe 
df1=df[0].str.split(',',expand=True)
df1
#its expand table whereeve ',' character present
```

```python
#find column heading
column_heading=soup.find_all('th')
column_heading
```

```python
#change column headings to string
all_header=[]
sss=str(column_heading)
clean_text=BeautifulSoup(sss,'lxml').get_text()
all_header.append(clean_text)
print(all_header)
```

```python
#set headers to dataframe
df2=pd.DataFrame(all_header)
df2
```

```python
#split the dataframe 
df3=df2[0].str.split(',',expand=True)
df3
#its expand table whereeve ',' character present
```

```python
#concat two data frames
frames=[df3,df1]
df4=pd.concat(frames)
df4.head()
```

```python
#change the column name as first row
df5=df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
#find the information and shape about recent d6ataframe
df5.info()
df5.shape
```

```python
#delete the None value rows from the dataframe
df6=df5.dropna(axis=0,how='any')
df6
```

```python
#delete first row in dataframe
df7=df6.drop(df6.index[0])
df7
```

```python
#change the column name as we want
df7.rename(columns={'[Place':'Place'},inplace=True)
df7.rename(columns={' Team]': 'Team'},inplace=True)
df7
```

```python
#remove ']' character from team column using strip
df7['Team']=df7['Team'].str.strip(']')
df7.head()
```
