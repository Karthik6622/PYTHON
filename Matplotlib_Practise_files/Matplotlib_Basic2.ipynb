# Matplotlib_Basic2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#here we are creating two samespace arrays using linspace attribute
#linspace attribute is used to create array with same space elemnents
x1=np.linspace(1,5,10)
print(x1)
y1=x1**5
y1
```

```python
#creating plot graph with linspace array values
plt.plot(y1,x1,color='r',lw=10,marker='*',mfc='yellow',ms=20)
plt.xlabel("KARTHIK")
plt.ylabel("GROWTH")
plt.title("Karthik Growth")
```

# Multiplots on same canvas

```python
#we can create multiple plots on the same canvas using Matplotlib, you can use the subplot or subplots function.
plt.subplot(2,1,1)#2-row,1-column,1-which place our graph should come
plt.plot(x1,y1)
plt.subplot(2,1,2)#2-row,1-column,2-which place our graph should come
plt.plot(y1,x1)
```

```python
#subplot
plt.subplot(1,2,2)#2-row,1-column,1-which place our graph should come
plt.plot(x1,y1)
plt.subplot(1,2,1)#2-row,1-column,2-which place our graph should come
plt.plot(y1,x1)
```

```python
#create 6 plots in one canvas using subplot
#1st_plot
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 1)
plt.plot(x,y)
#2nd_plot
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 2)
plt.plot(x,y)
#3rd_plot
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 3)
plt.plot(x,y)
#4th_plot
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 4)
plt.plot(x,y)
#5th_plot
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 5)
plt.plot(x,y)
#6th_plot
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()#this one used to show the plots.we can use show function to display plot
```

# Object Oriented Plotting

```python
#instantiating a plot object and then calling methods and attributes of that object to modify the plots properties.
#we use figure object for object oriented plottinng
#figure()
#A figure is the top-level container that holds all the elements of a plot. It can contain one or more subplots, axes, labels, legends, and other graphical elements.
fig=plt.figure()
axis=fig.add_axes([0,0,1,2])#left,bottom,length,height.whith this we can change Graph height,length as we need
axis.plot(x1,y1)
axis.set_xlabel("Karthik")
axis.set_ylabel("Growth")
axis.set_title("Karthik Growth")
```

```python
#how to change size of plot by axis values
fig=plt.figure()
axis=fig.add_axes([0.3,0.2,0.3,0.4])#left,bottom,length,height
axis.plot(x1,y1)
axis.set_xlabel("Karthik")
axis.set_ylabel("Growth")
axis.set_title("Karthik Growth")
```

```python
#how to display more than one plots using figure object
fig=plt.figure(figsize=(25,5))#length and height of figure
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)
ax1.plot(x1,y1)
ax2.plot(x1,y1)
plt.show()
```

```python
#how to display one graph into the another grapth using figure object
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,0.2])
ax1.plot(x1,y1,'r',marker='>')
ax1.set_xlabel("Karthik")
ax1.set_ylabel("Growth")
ax1.set_title("Karthik Growth")
ax2.plot(x,y,'y',marker='*',mfc='blue')
ax2.set_xlabel("Karthik",fontsize=20)
ax2.set_ylabel("Growth",fontsize=20)
ax2.set_title("Karthik Growth",fontsize=20)
```

```python
#subplots-create plot using subplot
fig,a=plt.subplots()
a.plot([56,4,3,2])
```

```python
#how to create two plots using subplots
fig,axes=plt.subplots(1,2)#1-row,2-column
axes[0].plot([56,4,3,2])
axes[1].plot([56,4,3,2])
axes[0].set_xlabel('X-axis (Left Subplot)')
axes[0].set_ylabel('Y-axis (Left Subplot)')
axes[1].set_xlabel('X-axis (Right Subplot)')
axes[1].set_ylabel('Y-axis (Right Subplot)')
axes[0].set_title('Left Subplot Title')
axes[1].set_title('Right Subplot Title')
fig.savefig('kk1.png',dpi=400)#here also we can save the plot using fig
```

```python
#tight_layout() function manage both plots not going to merge
#in above plot we can see second plot label touched first plot to avoid that we use tight_layout
fig,axes=plt.subplots(1,2)#1-row,2-column
axes[0].plot([56,4,3,2])
axes[1].plot([56,4,3,2])
axes[0].set_xlabel('X-axis (Left Subplot)')
axes[0].set_ylabel('Y-axis (Left Subplot)')
axes[1].set_xlabel('X-axis (Right Subplot)')
axes[1].set_ylabel('Y-axis (Right Subplot)')
axes[0].set_title('Left Subplot Title')
axes[1].set_title('Right Subplot Title')
fig.savefig('kk1.png',dpi=400)#here also we can save the plot using fig
plt.tight_layout()
```

# Customize Figures

```python
fig1=plt.figure()
aa=fig1.add_axes([0,0,1,1])
aa.plot(x1,y1,color='orange',linestyle='--',linewidth=10,marker='*',markersize=17,alpha=0.9,
markerfacecolor='yellow',markeredgecolor='blue',markeredgewidth=5
       )#alpha is use for transperency
```

```python
fig1=plt.figure()
aa=fig1.add_axes([0,0,1,1])
aa.plot(x1,y1,color='orange',ls='--',lw=10,marker='*',ms=17,alpha=0.9,
mfc='yellow',mec='blue',mew=5
       )#alpha is use for transperency
aa.set_xlim([0,3])#here we can set limit of x.it only display 0 to 3 values
aa.set_ylim([0,500])#here we can set limit of y.it only display 0 to 500 values
```

```python
# Scatter()-we can use the scatter() function to draw a scatter plot.
#create scatter plot by csv file data's
df=pd.read_csv('data_car_cross.csv',index_col=0)
df
```

```python
fig2=plt.figure()
a1=fig2.add_axes([0,0,1,1])
a1.set_xlabel('Horse Power')
a1.set_ylabel('Weight')
a1.set_title('Comparison')
a1.scatter(df['horse power'],df['weight'],color='r')
```

```python
#Histogram Plot-A histogram is a graph showing frequency distributions.
#and it's like a bar
plt.hist(df['cylinder'])
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
```

```python
#create plot,bar,piechart using csv file dataset
#plot
#here we finding total count of each cylinders and changing to dataframe
dd=df['cylinder'].value_counts()
#print(dd)
dd_to_dataframe=dd.reset_index()
plt.plot(dd_to_dataframe['index'],dd_to_dataframe['cylinder'])
plt.xlabel('Cylinders')
plt.ylabel('NO of Cylinders')
```

```python
#bar
#finding how much people bought car in every year
#using groupby find how much people bought car in each year 
d=df.groupby(['year'])['name'].count()
#changing groupby object to new Dataframe
dd=d.reset_index()
#adding columns to new dataframe
dd.columns=['year','name']
plt.bar(dd['year'],dd['name'])
plt.xlabel('Year')
plt.ylabel('No of car Buyers')
```

```python
#piechart
label1=df['cylinder'].unique()
print(label1)
plt.pie(df['cylinder'].unique(),labels=label1)
```

```python
#grid-we can use the grid() function to add grid lines to the plot.
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x, y)
plt.grid()
plt.show() 
```

```python
#how to create row and column grid
x1 = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y1 = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x1, y1)
plt.grid(axis='y')#axis=y-row grid,axis=x-column grid(x,y is not variables it's graph x and y position)
plt.show() 
```

```python
#create customized Grid
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x, y)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```
